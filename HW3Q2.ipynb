```
from pyspark import SparkContext
import pyspark.sql as sql
import random
```

```
#B
data = [random.randint(0,50) for _ in range(500)]

sc = SparkContext.getOrCreate()
rDD = sc.parallelize(data)
```

```
#C,D
sum = rDD.sum()
count = rDD.count()
mean = rDD.mean()
variance = rDD.variance()

print('C) ',sum,count,mean)
print('D) ',variance)
```

```
#E
counts = rDD.map(lambda x: (x,1))
counts = counts.reduceByKey(lambda x,y: x+y)
counts = counts.filter(lambda x: x[1]>10)
counts_list = counts.collect()

counts_list.sort(key=lambda x:x[0])

print('E) ',counts_list)
```
