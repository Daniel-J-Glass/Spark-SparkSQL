```
from pyspark import SparkContext
import random
```

```
#C
sc = SparkContext.getOrCreate()
rDD = sc.textFile('FileStore/tables/inputHW3.txt')

rDD = rDD.flatMap(lambda x: x.lower().split(' '))
```

```
#D
count = rDD.count()
print('D) ',count)

counts = rDD.map(lambda x: (x,1))\
            .reduceByKey(lambda x, y: x+y)

count_list = counts.collect()

print(count_list)
```

```
top = counts.takeOrdered(30, key=lambda x: -x[1])
print('E) ',top)
```
